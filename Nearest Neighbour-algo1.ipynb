```python
import numpy as np
import matplotlib.pyplot as plt

#sample training set
X_train=np.array([[1,1],[2,2.5],[3,1.2],[5.5,6.3],[6,9],[7,6]])
y_train=['red','red','red','blue','blue','blue']


#splitting the x_coordinates and y_coordinates as [0]and[1]:
plt.scatter(X_train[:,0],X_train[:,1],s=150,color=y_train)
plt.show()
```

```python
X_test=np.array([8,4])
plt.scatter(X_train[:,0],X_train[:,1],s=150,color=y_train)
plt.scatter(X_test[0],X_test[1],s=150,color='orange')
plt.show()
```

```python
#here x is a point in my training data and y 
#is the test data point  

def find_dist(x,y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
#calculating the distance b/w test data point and 
#all other training data points:and ultimately we chose the one with minimal distance
a=len(X_train)
computed_dist=np.zeros(a)
for i in range(a):
    computed_dist[i]=find_dist(X_train[i],X_test)
print(computed_dist)
    
```

```python
min_dist=np.argmin(computed_dist)
print(y_train[min_dist])
```

```python
#number of mistakes done in 1000i/p data:
a=len(X_train)
no_errors=0
computed_dist=np.zeros(a)
for j in range(1697,1797):
    X_test=digits.data[j]
    for i in range(a):
        computed_dist[i]=find_dist(X_train[i],X_test)
    min_dist=np.argmin(computed_dist)
    if y_train[min_dist]!=digits.target[j]:
        no_errors+=1
print(no_errors)
```
